# Emergency - 911 Calls

In this project I will do some basic analysis and visualisation on the **Emergency - 911 Calls** dataset from Kaggle. 

The dataset contains information on Fire, Traffic, EMS emergency calls for Montgomery County, PA over an 8 month period from December 2015 to August 2016.

##### About Montgomery County

> Montgomery County, officially the County of Montgomery, is a county in the U.S. state of Maryland. As of the 2010 census, the population was 971,777, with a 2014 estimate putting the population at 1,030,447. It is the most populous county in Maryland. The county seat and largest municipality is Rockville, although the census-designated place of Germantown is the most populous place.
    
##### About 911 Calls
    
> The three-digit telephone number "9-1-1" has been designated as the "Universal Emergency Number," for citizens throughout the United States to request emergency assistance. It is intended as a nationwide telephone number and gives the public fast and easy access to a Public Safety Answering Point (PSAP).

|Feature|Description|Data type|
|---|---|---|
|lat |  Latitude|Float variable|
|lng| Longitude|Float variable|
|desc| Description of the Emergency Call|String variable|
|zip| Zipcode|Float variable|
|title| Title|String variable|
|timeStamp| YYYY-MM-DD HH:MM:SS|String variable|

|addr| Address|Integer variable|
|e| Dummy variable (always 1)|String variable|


```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (6, 4)
```

```python
#Read in the data
df = pd.read_csv('Resources/911.csv')
```

```python
df.head()
```

```python
df.info()
```

From looking at the dataframe I can see that I need to -

##### Clean
- Convert datatype of the **timeStamp** feature from a string to datetime

##### Feature creation
- Creat category and subcategory features based on the **title** feature
- Creae the Year, Month and Day of Week features from the **timeStamp** feature

```python
# Convert timeStamp to datetime
df.timeStamp = pd.to_datetime(df.timeStamp)
```

```python
# Create category and subcategory features
df['category'] = df.title.apply(lambda x: x.split(':')[0])
df['subcategory'] = df.title.apply(lambda x: x.split(':')[1])
```

```python
df.head(3)
```

## Analysis

Let's look out the top 5 zipcodes and townships that the calls originated.

```python
df.zip.value_counts().head()
```

```python
df.twp.value_counts().head()
```

For ~100k entries, let's see how many unique call titles there were?

```python
df.title.nunique()
```

Let's look at the reasons for the calls recieved.

```python
df.category.value_counts()
```

```python
sns.countplot(df.category)
plt.title('Emergency call catagories\n')
plt.show()
```

Next let's seperate out the timestamp to date, month, weekday and hour.

```python
df['hour'] = df['timeStamp'].dt.hour
df['month'] = df['timeStamp'].dt.month
df['week_day'] = df['timeStamp'].dt.dayofweek
df['date'] = df['timeStamp'].dt.date
```

```python
df.head(2)
```

The weekday is an integer, ranging from 0 - 6, and it might not be instantly clear which number refers to which day of the week. I will map the integer to a Mon-Sun string.

```python
days = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df.week_day = df.week_day.map(days)
```

```python
df.head(3)
```

Let's combine the newly created features, to check out the most common call reasons based on the day of the week.

```python
sns.countplot(df.week_day, hue=df.category)
plt.legend(bbox_to_anchor=(1.25,1))
plt.title('Number of daily calls for each catagory\n')
plt.show()
```

It makes sense for the number of traffic related 911 calls to be the lowest during the weekends, what's also iteresting is that Emergency Service related calls are also low during the weekend.

Now, let's check out the relationship between the number of calls and the month.

```python
sns.countplot(df.month, hue=df.category)
plt.title('Number of monthly calls for each catagory\n')
plt.legend(bbox_to_anchor=(1.25,1))
plt.show()
```

```python
by_month = pd.groupby(df, by='month').count()
```

```python
by_month['e'].plot.line(y='e')
plt.title('Number of calls per month\n')
plt.ylabel('Number of calls')
plt.xlabel('Month')
plt.show()
```

Next let's fit the number of calls to a month and see if there's any definitive correlation between the two.

```python
by_month.reset_index(inplace=True)
```

```python
sns.lmplot(x='month', y='e', data=by_month)
plt.title('Number of calls per month\n')
plt.ylabel('Number of calls')
plt.xlabel('Month')
plt.show()
```

From the above it seems that there are fewer emergency calls during the holiday seasons.

Next let's group and plot the relationship between the number of calls and the date.

```python
pd.groupby(df,'date').count()['e'].plot.line(y='e')
plt.legend().remove()
plt.title('Calls per day\n')
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

We can also check out the same plot for each call catagory separately.

```python
pd.groupby(df[df['category']=='Traffic'],'date').count().plot.line(y='e')
plt.legend().remove()
plt.title('Traffic\n')
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
pd.groupby(df[df['category']=='Fire'],'date').count().plot.line(y='e')
plt.legend().remove()
plt.title('Fire\n')
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
pd.groupby(df[df['category']=='EMS'],'date').count().plot.line(y='e')
plt.legend().remove()
plt.title('EMS\n')
plt.ylabel('Number of calls')
plt.xlabel('Date')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

Next we'll create a heatmap for the counts of calls on each hour, during a given day of the week.

```python
day_hour = df.pivot_table(values='lat',index='week_day',columns='hour',aggfunc='count')
day_hour
```

Now using the newly created dataframe we'll plot a heatmap.

```python
sns.heatmap(day_hour)
plt.tight_layout()
```

We see that most calls take place around the end of office hours on weekdays. 

We can create a clustermap to pair up similar Hours and Days.

```python
sns.clustermap(day_hour)
plt.show()
```
