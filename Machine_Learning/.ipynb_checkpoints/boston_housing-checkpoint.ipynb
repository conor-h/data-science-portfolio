This dataset was originally completed as part of the Data Science Immersive course at General Assembly.

In this project, I will evauate the performance and predictive power of a model that has been trained and tested on data collected from homes in suburbs of Boston, Massachusetts. A model trained on this data that is seen as a good fit could then be used to make certain predictions about a home â€” in particular, its monetary value.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
sns.set_style("whitegrid")
```

```python
# Load the dataset
data = pd.read_csv('Resources/housing.csv', index_col=0)
```

```python
data.head()
```

Having worked on the dataset previously, the following data cleaning steps will be made to the dataset:

- 16 data points have an 'MEDV' value of 50.0. These data points likely contain missing or censored values and will be removed.
- 1 data point has an 'RM' value of 8.78. This data point can be considered an outlier and will be removed.
- The features 'RM', 'LSTAT', 'PTRATIO', and 'MEDV' are essential. The remaining non-relevant features will be dropped.
- The feature 'MEDV' will be multiplicatively scaled to account for market inflation.

```python
# Remove rows where the MEDV value is 50.0
data = data[data.MEDV != 50.0]

# Remove row where RM value is 8.78
data = data[data.RM != 8.78]

# Drop non-relevant features
data = data.drop(['CRIM','ZN','INDUS','CHAS','NOX','AGE','DIS','RAD','TAX','B'], axis=1)

# Multiply MEDV to convert to full dollar value at current market values
data['MEDV'] = data['MEDV'] * 21000
```

```python
# Define the features and target variable
prices = data['MEDV']
features = data.drop('MEDV', axis=1)
```

## Building a Model

In this section, I will develop the tools and techniques necessary for a model to make a prediction. Being able to make accurate evaluations of each model's performance through the use of these tools and techniques helps to greatly reinforce the confidence in their predictions.

```python
# Import libraries necessary for modelling
from sklearn.metrics import r2_score, make_scorer
from sklearn.cross_validation import ShuffleSplit, train_test_split
import sklearn.learning_curve as curves
from sklearn.tree import DecisionTreeRegressor
from sklearn.grid_search import GridSearchCV
```

### Define a Performance Metric

For this project, I will calculate the coefficient of determination, R<sup>2</sup>, to quantify my model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions.

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the mean of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**.

I will define a function that returns the R<sup>2</sup> score for given true and predicted data next.

```python
# Define function to return the R2 score
def performance_metric(y_true, y_predict):
    
    score = r2_score(y_true,y_predict)

    return score
```

### Goodness of Fit

To test the function, assume that a dataset contains five data points and a model made the following predictions for the target variable:

|True Value	|Prediction|
|---|---|
|3.0	|2.5|
|-0.5	|0.0|
|2.0	|2.1|
|7.0	|7.8|
|4.2	|5.3|

Now I'll use the performance_metric function and calculate this model's R<sup>2</sup> score.

```python
y_true = [3, -0.5, 2, 7, 4.2]
y_pred = [2.5, 0.0, 2.1, 7.8, 5.3]
print 'The model has a R2 score of', performance_metric(y_true, y_pred)
```

### Splitting the Data

I will split the data into training and test sets.

```python
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.3, random_state=99)
```

Splitting the data into training and test sets allows for the model to train on the training set and then feed in the unseen test set to evaluate the performance of the model.

Training and testing on the same data will not give us a genuine evaluation of the model, as it has already been testing the data when training, and thus might not perform well in real-world scenarios with truely unseen data. This can be the cause of "overfitting", i.e. the model will perform well at making predictions on the training data, but perform poorly on test data.

### Model Performance

In this section, I will calculate the performance of several models with varying sizes of training data. The training and test scores for each model will then be plotted.

```python
# Create 10 cross-validation sets for training and testing
cv = ShuffleSplit(features.shape[0], n_iter = 10, test_size = 0.2, random_state = 0)

# Generate the training set sizes increasing by 50
train_sizes = np.rint(np.linspace(1, features.shape[0]*0.8 - 1, 9)).astype(int)

# Create the figure window
fig = plt.figure(figsize=(10,7))

# Create three different models based on max_depth
for k, depth in enumerate([1,3,6,10]):

    # Create a Decision tree regressor at max_depth = depth
    regressor = DecisionTreeRegressor(max_depth = depth)

    # Calculate the training and testing scores
    sizes, train_scores, test_scores = curves.learning_curve(
        regressor, features, prices, cv = cv, train_sizes = train_sizes, scoring = 'r2')

    # Find the mean and standard deviation for smoothing
    train_std = np.std(train_scores, axis = 1)
    train_mean = np.mean(train_scores, axis = 1)
    test_std = np.std(test_scores, axis = 1)
    test_mean = np.mean(test_scores, axis = 1)

    # Subplot the learning curve
    ax = fig.add_subplot(2, 2, k+1)
    ax.plot(sizes, train_mean, 'o-', color = 'r', label = 'Training Score')
    ax.plot(sizes, test_mean, 'o-', color = 'g', label = 'Test Score')
    ax.fill_between(sizes, train_mean - train_std, 
                    train_mean + train_std, alpha = 0.15, color = 'r')
    ax.fill_between(sizes, test_mean - test_std, 
                    test_mean + test_std, alpha = 0.15, color = 'g')

    # Define the labels
    ax.set_title('max_depth = %s'%(depth))
    ax.set_xlabel('Number of Training Points')
    ax.set_ylabel('Score')
    ax.set_xlim([0, features.shape[0]*0.8])
    ax.set_ylim([-0.05, 1.05])

# Visual aesthetics
ax.legend(bbox_to_anchor=(1.05, 2.05), loc='lower left', borderaxespad = 0.)
fig.suptitle('Decision Tree Regressor Learning Performances\n', fontsize = 16, y = 1.03)
fig.tight_layout()
fig.show()
```

The four graphs above show learning curves for a decision tree model with different maximum depths (1, 3, 6 & 10). Each graph visualises the learning curves for both training and test as the size of the training set is increased. The shaded region of a learning curve denotes the uncertainty of that curve (measured as the standard deviation). The model is scored on both the training and test sets using the R<sup>2</sup> score.

Looking at the learning curve for the model with a max_depth = 3: 

- The score of the training curve decreases as more training points are added. This occurs because with fewer training points, the model can modify its paramters to better approximate the targets; but as the number of training points increases, perfectly fitting them becomes more difficult, and the training score goes down.
- The score of the test curve increases as more training points are added, but there's a slight dip after adding more than 350 training points. The score starts lower because the model has not yet learned enough to predict test points. As the model receives more training points, and hence more information, it is better suited to predict unseen data.
- The training and testing curve seem to be converging to a score of ~0.8. This usually happens when the model has stretched its limits of extracting information from the training data even though more training points are being added. So the score stabilises.

Therefore, having more training points might not benefit the model (with max_depth = 3).

### Model Complexity

In this section, I will calculate the performance of the model as the model complexity increases. The learning and testing errors rates will then be plotted.

```python
# Create 10 cross-validation sets for training and testing
cv = ShuffleSplit(X_train.shape[0], n_iter = 10, test_size = 0.2, random_state = 0)

# Vary the max_depth parameter from 1 to 10
max_depth = np.arange(1,11)

# Calculate the training and testing scores
train_scores, test_scores = curves.validation_curve(
    DecisionTreeRegressor(), X_train, y_train, param_name = "max_depth", 
    param_range = max_depth, cv = cv, scoring = 'r2')

# Find the mean and standard deviation for smoothing
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the validation curve
plt.figure(figsize=(7, 5))
plt.title('Decision Tree Regressor Complexity Performance\n')
plt.plot(max_depth, train_mean, 'o-', color = 'r', label = 'Training Score')
plt.plot(max_depth, test_mean, 'o-', color = 'g', label = 'Test Score')
plt.fill_between(max_depth, train_mean - train_std, 
                train_mean + train_std, alpha = 0.15, color = 'r')
plt.fill_between(max_depth, test_mean - test_std, 
                test_mean + test_std, alpha = 0.15, color = 'g')

# Visual aesthetics
plt.legend(loc = 'lower right')
plt.xlabel('Maximum Depth')
plt.ylabel('Score')
plt.ylim([-0.05,1.05])
plt.show()
```

The graph above shows a decision tree model that has been trained and tested on the training data using different maximum depths. The graph produced two complexity curves â€” one for training and one for testing. Similar to the learning curves previously, the shaded regions of both the complexity curves denote the uncertainty in those curves, and the model is scored on both the training and rest sets using the R<sup>2</sup> score.

#### Bias-Variance Tradeoff

- The model trained with max_depth = 1: The model suffers from high bias at this depth. Looking at the graph, we can see that both training and validation scores are low, and similar we can say that the model is over-simplified and is not capturing the underlying relationships present in the data for both training and test sets.
- The model trained with max_depth = 10: The model suffers from high variance at this depth, and is overfitting on the training data. In the graph, it can see that the training score at this depth is almost equal to 1.0, while the test score is lower, at ~0.7 and is decreasing. The curves are diverging away from each other at this point as the model max_depth increases.

From the graph above, a maximum depth of 4 results in a model that best generalises to unseen  test data. That depth is the sweet spot for model complexity, as the model performs similar on training and test data, while the overall score for both is still relatively high at between 0.7 to 0.8. A depth lower than that gives us poor training and test score, while a higher depth overfits on the training data, leading to a lower test score.

## Evaluating Model Performance

In this final section, I will construct a model and make predictions on a set of sample houses.

I will use **Grid Search** and **Cross Validation** to determine the optimal parameters for model.

### Grid Search

Grid search is an approach for parameter tuning that will methodically build and evaluate a model for each combination of algorithm parameters specified in a grid.

In order to optimize my model, I will apply grid search on a number of the parameters. I will provide a list of possible values for those parameters. The grid search will then return the best parameter values for my model, after fitting the training dataset.

### Cross Validation

Cross-validation is another method to estimate the skill of a method on unseen data, similar to using a train-test split. Cross-validation systematically creates and evaluates multiple models on multiple subsets of the dataset, it does this by splitting the dataset into k partitions of equal size, and then running k separate learning experiments on the training data.

This method is benefitial when using grid search to optimize a model because it allows us to identify parameter settings that perform well for different test sets. If we had a single testing set, it is easy to tune a model to perform well for that specific test set (and result in overfitting on the training set in this case), while cross validation allows us to generalise the results.

### Fitting a Model

I will define a function to perform a grid search over the **'max_depth'** parameter for a  decision tree regressor trained on the input data (X, y).

```python
def fit_model(X, y):
    
    # Create cross-validation sets
    cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.20, random_state = 0)

    # Define the model
    regressor = DecisionTreeRegressor()

    # Identify the parameter range
    params = {'max_depth': range(1,11)}

    # Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring = make_scorer(performance_metric)

    # Define the grid search with the parameters
    grid = GridSearchCV(regressor, params, scoring, cv=cv_sets)

    # Fit the grid search object to the input data
    grid = grid.fit(X, y)

    # Return the optimal model
    return grid.best_estimator_
```

### Optimal Model

I will use the function on the training dataset, this will perform a grid search over the 'max_depth' parameter for a decision tree regressor and return the optimal value.

```python
# Fit the training data to the model
reg = fit_model(X_train, y_train)

print 'The optimal "max_depth" was found to be', reg.get_params()['max_depth'],'.'
```

The optimal model was found to have a **max_depth** of 4, this is the same as what I identified previously when looking at the model performance.

### Sensitivity

An optimal model is not necessarily a robust model. Sometimes, a model is either too complex or too simple to sufficiently generalize to new data. Sometimes, a model could use a learning algorithm that is not appropriate for the structure of the data given. Other times, the data itself could be too noisy or contain too few samples to allow a model to adequately capture the target variable â€” i.e., the model is underfitted.

I will define a function to run the **fit_model** function ten times with different training and testing sets to see how the prediction for a specific house changes with respect to the data it is trained on.

```python
def PredictTrials(X, y, fitter, data, house):

    # Create an empty list to store the predicted prices
    prices = []

    for k in range(10):
        # Split the data
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = k)

        # Fit the model
        reg = fitter(X_train, y_train)

        # Make preditions
        pred = reg.predict([data[house-1]])[0]
        #Append the prices to the empty list
        prices.append(pred)

        print "Trial {}: ${:,.2f}".format(k+1, pred)

    # Print the price range
    print "\nRange in prices: ${:,.2f}".format(max(prices) - min(prices))
```

```python
# Find the price range for House 1
PredictTrials(features, prices, fit_model, house_data, 1)
```

The range in the selling price for House 1 was found to be \$97,800.83.
